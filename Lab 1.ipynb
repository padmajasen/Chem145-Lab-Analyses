```python
import numpy as np
import matplotlib.pyplot as plt

line_10 = np.loadtxt('10uM data.txt')
line_15 = np.loadtxt('15uM data.txt')
line_20 = np.loadtxt('20uM data.txt')
line_25 = np.loadtxt('25uM data.txt')

ax = plt.axes()
ax.set_xlim([400,600])
ax.plot(line_10[:,0], line_10[:,1],label='10')
ax.plot(line_15[:,0], line_15[:,1],label='15')
ax.plot(line_20[:,0], line_20[:,1],label='20')
ax.plot(line_25[:,0], line_25[:,1],label='25')

plt.xlabel('wavelength (nm)')
plt.ylabel('Absorbance')
plt.title('Absorbance vs Wavelength for the Fluroscein Main Dilutions Series')
ax.legend();
```
